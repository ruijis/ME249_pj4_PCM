```python
surveyID = 10
pd.set_option('display.max_rows', 10)
df = dF.select_Columns_fromCSV(filepath, columns, keyColumn, surveyID)
# change chiar.staus column data to chiarStatus
df.columns = ['Tout', 'back_fan', 'back_heater', 'seat_fan', 'seat_heater', 'Ta_chair', 'RH_chair', 'survey.id', 'Preference', 'chairStatus']
df
```

## Transfering preference data strings to numerical values

```python
df['Preference'] = np.where((df.Preference == 'Warmer'), 1, df.Preference)
df['Preference'] = np.where((df.Preference == 'No change'), 2, df.Preference)
df['Preference'] = np.where((df.Preference == 'Cooler'), 3, df.Preference)
df['chairStatus'] = np.where((df.chairStatus == 'No'), 0, df.chairStatus)
df['chairStatus'] = np.where((df.chairStatus == 'Yes'), 1, df.chairStatus)
df
```

## Keeping the data when chairStatus is 1, otherwise deleting the data

```python
final_df = df.loc[df['chairStatus'] > 0]
final_df
```

## Summing all fan values and all heater values

```python
fan_list = ['back_fan', 'seat_fan']
heater_list = ['back_heater',  'seat_heater']
# sum column list accross row
final_df['fan'] = final_df[fan_list].sum(axis=1)
final_df['heater'] = final_df[heater_list].sum(axis=1)
io_df = final_df[['Tout', 'Ta_chair', 'RH_chair', 'fan', 'heater', 'Preference']]
io_df
```

### integrate fan and heater and replace reference

```python
io_df['heater'] = io_df['heater']*(-1)
io_df['prefer'] = io_df['fan'] + io_df['heater']
io_df['prefer'] = io_df['prefer']/200 + 2
# 1-2: prefer warmer (feeling cold); 2: no change; 2-3: prefer cooler (feeling hot).
integrated_df = io_df[['Tout', 'Ta_chair', 'RH_chair', 'prefer']]
integrated_df
```

## Simulating and transfering PMV values to fan or heat strips control signals as inputs and thermal preferences as outputs
<!-- TSV	-3	-2	-1	0	1	2	3
fan	0	0	0	0	50,100	100,150	150,200
heater	150,200	100,150	50,100	0	0	0	0
Preference	3	3	1	1	1	2	2 -->

```python
import random
from pythermalcomfort.models import pmv_ppd
from pythermalcomfort.psychrometrics import v_relative

tdb = 25
tr = 25.0
v = 0.15
rh = 50.0
met = 1.0
clo = 1.0
vr = v_relative(v, met)
standard= "ashrae"

for i in range(200):
    tout = random.uniform(10.0, 20.0)
    tdb = random.uniform(15.0, 30.0)
    tr = (tout + tdb)/2
    rh = random.uniform(30.0, 70.0)
    pmv = pmv_ppd(tdb=tdb, tr=tr, vr=vr, rh=rh, met=met, clo=clo, standard=standard)["pmv"]
    # for pmv between 0 and 3, give corresponding values
    if 0 < pmv <= 1:
        sim_P = 1
        sim_fan = random.randint(50.0, 100.0)
        sim_heater = 0
    if 1 < pmv <= 2:
        sim_P = 2
        sim_fan = random.randint(100.0, 150.0)
        sim_heater = 0
    if 2 < pmv <= 3:
        sim_P = 2
        sim_fan = random.randint(100.0, 150.0)
        sim_heater = 0
    # for pmv between 0 and -3, give corresponding values
    if -1 < pmv <= 0:
        sim_P = 1
        sim_fan = 0
        sim_heater = random.randint(50.0, 100.0)
    if -2 < pmv <= -1:
        sim_P = 3
        sim_fan = 0 
        sim_heater = random.randint(100.0, 150.0)
    if -3 < pmv <= -2:
        sim_P = 3
        sim_fan = 0
        sim_heater = random.randint(100.0, 150.0)
    new_row = {'Tout': tout, 'Ta_chair': tdb, 'RH_chair': rh, 'fan': sim_fan, 'heater': sim_heater, 'Preference': sim_P}
    #append row to the dataframe
    io_df = io_df.append(new_row, ignore_index=True)
```
